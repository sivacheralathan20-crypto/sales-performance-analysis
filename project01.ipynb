```python
# Convert dates with day-first format
df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True, errors='coerce')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], dayfirst=True, errors='coerce')

# Check for failed conversions
print(df[['Order_Date', 'Ship_Date']].isnull().sum())
```

```python
# Convert using exact format
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%Y-%m-%d')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], format='%Y-%m-%d')

# Create new columns
df['Shipping_Days'] = (df['Ship_Date'] - df['Order_Date']).dt.days
df['Month'] = df['Order_Date'].dt.month
df['Year'] = df['Order_Date'].dt.year

# Verify
df.info()
df.head()
```

```python
#“A relatively small group of high-value customers contributes nearly half of the company’s total revenue, indicating strong customer concentration and the importance of customer retention strategies.”
# Customer-wise total sales
customer_sales = df.groupby('Customer_Name')['Sales'].sum().sort_values(ascending=False)

# Top 20% customers
top_20_count = int(len(customer_sales) * 0.2)
top_customers = customer_sales.head(top_20_count)

# Contribution percentage
total_sales = customer_sales.sum()
top_sales = top_customers.sum()
contribution = (top_sales / total_sales) * 100

print("Top 20% customers contribute:", round(contribution, 2), "% of total sales")
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='mixed', dayfirst=True)
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], format='mixed', dayfirst=True)

df['Shipping_Days'] = (df['Ship_Date'] - df['Order_Date']).dt.days
df['Month'] = df['Order_Date'].dt.month
df['Year'] = df['Order_Date'].dt.year
```

```python
# Category-wise sales
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print("Category Sales:")
print(category_sales)

print("\n")

# Sub-category-wise sales
sub_category_sales = df.groupby('Sub_Category')['Sales'].sum().sort_values(ascending=False)
print("Sub-Category Sales:")
print(sub_category_sales.head(10))
```

“Technology is the strongest-performing category, driven primarily by high-demand products such as Phones and Accessories. Furniture also shows strong performance, with Chairs being a major revenue contributor.”

Strategic recommendations:

Increase focus on Phones
Cross-sell Accessories
Expand high-performing Furniture lines

```python
# Region-wise sales
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

# Percentage contribution
region_percentage = (region_sales / region_sales.sum()) * 100

print("Region-wise Sales:")
print(region_sales)

print("\nPercentage Contribution:")
print(region_percentage.round(2))
```

The company maintains a fairly diversified regional revenue structure, though West and East dominate with over 60% combined sales. South shows the lowest contribution, highlighting a potential opportunity for market expansion.”

Strategic recommendation:
Increase marketing in South
Analyze why South underperforms
Protect high-performing West and East customer base

```python
# Segment analysis
segment_sales = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
segment_orders = df.groupby('Segment')['Order_ID'].count()
segment_avg_order = df.groupby('Segment')['Sales'].mean().sort_values(ascending=False)

print("Total Sales by Segment:")
print(segment_sales)

print("\nTotal Orders by Segment:")
print(segment_orders)

print("\nAverage Order Value by Segment:")
print(segment_avg_order.round(2))
```

```python
import pandas as pd

df = pd.read_csv("data/superstore_final_dataset (1).csv", encoding="latin1")

print("Dataset loaded successfully")
```

```python
# Convert mixed date formats
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='mixed', dayfirst=True)
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], format='mixed', dayfirst=True)

# Create shipping days column
df['Shipping_Days'] = (df['Ship_Date'] - df['Order_Date']).dt.days

# Check
print(df[['Order_Date', 'Ship_Date', 'Shipping_Days']].head())
```

```python
shipping_analysis = df.groupby('Ship_Mode')['Shipping_Days'].mean().sort_values()

sales_by_ship = df.groupby('Ship_Mode')['Sales'].mean().sort_values(ascending=False)

print("Average Shipping Days:")
print(shipping_analysis.round(2))

print("\nAverage Sales by Ship Mode:")
print(sales_by_ship.round(2))
```

```python
# Create Month column from Order_Date
df['Month'] = df['Order_Date'].dt.month

# Check
print(df[['Order_Date', 'Month']].head())
```

```python
# Monthly sales trend
monthly_sales = df.groupby('Month')['Sales'].sum().sort_values(ascending=False)

print("Monthly Sales:")
print(monthly_sales)
```

```python
# Top 10 products contribution
product_sales = df.groupby('Product_Name')['Sales'].sum().sort_values(ascending=False)

top_10_sales = product_sales.head(10).sum()
total_sales = product_sales.sum()

contribution = (top_10_sales / total_sales) * 100

print("Top 10 products contribute:", round(contribution, 2), "% of total sales")
```

```python
import matplotlib.pyplot as plt

region_sales.plot(kind='bar')
plt.title('Region-wise Sales')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Top 10 sub-categories
top_sub = df.groupby('Sub_Category')['Sales'].sum().sort_values(ascending=False).head(10)

# Plot
top_sub.plot(kind='bar')

plt.title('Top 10 Sub-Categories by Sales')
plt.xlabel('Sub Category')
plt.ylabel('Sales')
plt.xticks(rotation=45)

plt.show()
```
